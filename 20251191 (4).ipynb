#16
스택을 사용하는 위상 정렬 알고리즘은 진입 차수(들어오는 화살표 수)를 기록한 배열을 관리하며 작동합니다. 진입 차수가 0이 된 정점들을 스택에 넣어가며 탐색하는 방식입니다.

```
topologicalSort(G):
    // 1. 모든 정점의 진입 차수 계산하기
    for each vertex u in G:
        in_degree[u] = 0
    for each vertex u in G:
        for each neighbor v of u:
            in_degree[v] = in_degree[v] + 1
            
    // 2. 진입 차수가 0인 정점들을 스택에 넣기
    create an empty stack S
    for each vertex u in G:
        if in_degree[u] == 0:
            S.push(u)
            
    // 3. 위상 정렬 결과를 담을 리스트
    create an empty list result
    
    // 4. 스택이 빌 때까지 반복
    while S is not empty:
        u = S.pop()          // 스택에서 정점을 하나 꺼냄
        result.append(u)     // 결과 리스트에 추가
        
        // 꺼낸 정점과 연결된 모든 이웃 정점의 진입 차수 감소시키기
        for each neighbor v of u:
            in_degree[v] = in_degree[v] - 1
            // 새롭게 진입 차수가 0이 된 정점이 있다면 스택에 푸시
            if in_degree[v] == 0:
                S.push(v)
                
    // 5. 사이클 존재 여부 체크 (정렬된 정점 수가 전체 정점 수와 다른 경우)
    if length(result) != length(G.vertices):
        return "Graph has a cycle (Topological sort impossible)"
    else:
        return result
```

A → B : 2
A → C : 5
C → B : -4

시작 정점을 A로 하면,

다익스트라는 A→B의 비용 2를 최단거리로 먼저 확정한다.
그러나 실제 최단경로는 A→C→B 이며 비용은 5 + (-4) = 1 이다.

즉, 음의 가중치 때문에 이미 확정된 정점의 거리가 더 작아질 수 있으므로 다익스트라 알고리즘은 올바른 결과를 보장하지 못한다.
